## Soal 1: Series - Harga Produk

Seorang analis ingin menyimpan daftar harga produk dalam format Pandas Series agar lebih mudah dianalisis.

Lengkapi kode berikut agar:

1.  Membuat `Series` berisi harga produk: `[12000, 15000, 18000, 21000]`
2.  Memberikan label (index) berupa nama produk: `["Sabun", "Shampoo", "Pasta Gigi", "Sikat Gigi"]`
3.  Menyimpan `Series` dengan nama `harga_produk`
4.  Menampilkan seluruh isi `Series`, nama produk dengan harga tertinggi, dan rata-rata harga.

Hint 1: Harga tertinggi gunakan `.idxmax()`

Hint 2: Harga rata-rata gunakan `.mean()`


```python
import pandas as pd

# TODO: Buat Series harga_produk dari list harga dan index nama produk
harga_produk = pd.Series([12000,15000,18000,21000])
harga_produk.index = ["Sabun", "Shampoo", "Pasta Gigi", "Sikat Gigi"]

# Tampilkan seluruh isi Series
print(harga_produk)
```

```python
# TODO: Tampilkan nama produk dengan harga tertinggi
harga_produk.idxmax()
```

```python
# TODO: Hitung dan tampilkan rata-rata harga
harga_produk.mean()
```

## Soal 2: DataFrame - Data Siswa

Seorang guru ingin menyimpan data siswa dalam bentuk tabel untuk menganalisis performa mereka.

Lengkapi kode berikut agar:

1.  Membuat DataFrame `df_siswa` dari data berikut:
    -   `Nama`: `["Adit", "Rina", "Budi", "Sari"]`
    -   `Kelas`: `["10A", "10A", "10B", "10B"]`
    -   `Nilai`: `[90, 85, 78, 88]`
2.  Menampilkan seluruh DataFrame
3.  Filter Siswa dari kelas A
4.  Menampilkan daftar siswa dengan nilai > 80

```python
import pandas as pd

# TODO: Buat DataFrame dari dictionary
df_siswa = pd.DataFrame({'Nama':["Adit", "Rina", "Budi", "Sari"],
                         'Kelas':["10A", "10A", "10B", "10B"],
                         'Nilai':[90, 85, 78, 88]})

# Tampilkan seluruh data
print(df_siswa)

# TODO: Filter siswa kelas A
df_siswa[df_siswa['Kelas'] == '10A']
```

```python
# TODO: Tampilkan data siswa yang memiliki nilai di atas 80
df_siswa[df_siswa['Nilai'] > 80]
```

## Soal 3: read_csv - Data Penjualan

Anda diberikan file CSV berjudul `data_penjualan.csv` dengan struktur seperti ini:

| Tanggal    | Produk     | Jumlah | Harga |     |
| ---------- | ---------- | ------ | ----- | --- |
| 2024-01-01 | Sabun      | 20     | 3000  |     |
| 2024-01-01 | Pasta Gigi | 15     | 7000  |     |
| 2024-01-02 | Sabun      | 25     | 3000  |     |
| 2024-01-02 | Sikat Gigi | 10     | 5000  |     |

1. Baca dan simpan file CSV ke dalam variable `df`
2. Cetak DataFrame dan cetak info dari DataFrame
3. Ubah tipe data kolom `Tanggal` menjadi tipe `datetime`
4. Ubah kolom `Tanggal` menjadi index

```python
import pandas as pd

# TODO: Baca file CSV
df = pd.read_csv('data_penjualan.csv')

# TODO: Cetak DataFrame
print(df)

# TODO: Ubah tipe data kolom Tanggal menjadi datetime
df['Tanggal'] = pd.to_datetime(df['Tanggal'])

# TODO: Ubah kolom Tanggal menjadi index
df.set_index('Tanggal', inplace=True)
```
